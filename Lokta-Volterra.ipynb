```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

global alpha, beta, gamma, delta
alpha = 4
beta = 2
gamma = 3
delta = 1
```

### Funções f

```python
def f1(a, b, y1j, y2j, t): #lado direito da EDO
    return (a - b * y2j) * y1j

def f2(c, d, y1j, y2j, t):
    return (-c + d * y1j) * y2j
```

### Runge-Kutta

```python
def rk4(h):
    x = np.arange(0, 5+h, h)
    y1 = np.empty_like(x)
    y2 = np.empty_like(x)
    y1[0] = 3
    y2[0] = 5
    for j in range(len(x)-1):
        k1_1 = (alpha - beta * y2[j]) * y1[j]
        k2_1 = (alpha - beta * (y2[j] + h/2 * k1_1)) * (y1[j] + h/2 * k1_1)
        k3_1 = (alpha - beta * (y2[j] + h/2 * k2_1)) * (y1[j] + h/2 * k2_1)
        k4_1 = (alpha - beta * (y2[j] + h * k3_1)) * (y1[j] + h * k3_1)

        y1[j+1] = y1[j] + h/6 * (k1_1 + 2*k2_1 + 2*k3_1 + k4_1)

        k1_2 = (-gamma + delta * y1[j]) * y2[j]
        k2_2 = (-gamma + delta * (y1[j] + h/2 * k1_2)) * (y2[j] + h/2 * k1_2)
        k3_2 = (-gamma + delta * (y1[j] + h/2 * k2_2)) * (y2[j] + h/2 * k2_2)
        k4_2 = (-gamma + delta * (y1[j] + h * k3_2)) * (y2[j] + h * k3_2)

        y2[j+1] = y2[j] + h/6 * (k1_2 + 2*k2_2 + 2*k3_2 + k4_2)

    return x, y1, y2
```

### Time plot

```python
h1, h2, h3 = 0.001, 0.0005, 0.00025

figure1, (ax1, ax2, ax3) = plt.subplots(1, 3)
figure1.set_figwidth(18.0)
figure1.set_figheight(6.0)

figure1.suptitle(r"Solução do sistema de Lokta-Volterra")

ax1.set_title(r"$h = 0.001$")
ax2.set_title(r"$h = 0.0005$")
ax3.set_title(r"$h = 0.00025$")


ax1.plot(rk4(h1)[0], rk4(h1)[1], label = "coelhos", color = 'C1', linestyle='-', linewidth=2)
ax1.plot(rk4(h1)[0], rk4(h1)[2], label = "raposas", color = 'C2', linestyle='-', linewidth=2)
ax1.legend()

ax2.plot(rk4(h2)[0], rk4(h2)[1], label = "coelhos", color = 'C1', linestyle='-', linewidth=2)
ax2.plot(rk4(h2)[0], rk4(h2)[2], label = "raposas", color = 'C2', linestyle='-', linewidth=2)
ax2.legend()

ax3.plot(rk4(h3)[0], rk4(h3)[1], label = "coelhos", color = 'C1', linestyle='-', linewidth=2)
ax3.plot(rk4(h3)[0], rk4(h3)[2], label = "raposas", color = 'C2', linestyle='-', linewidth=2)
ax3.legend()

x_values = np.linspace(0, 5, 30)
```
